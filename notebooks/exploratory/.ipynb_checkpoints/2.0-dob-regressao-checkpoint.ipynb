```python
import pandas as pd
import numpy as np
```

## Criando dataset para modelagem

```python
gee = pd.read_csv("../../data/processed/gee_coleta.csv")
gee["week"] = pd.to_datetime(gee["date"]).dt.to_period("W")

vigi = pd.read_csv("../../data/processed/ciano_vigi.csv")
vigi["week"] = pd.to_datetime(vigi["Data da coleta"]).dt.to_period("W")
vigi_selected = vigi.loc[vigi["Nome da ETA / UTA"] == "MOINHOS DE VENTO",
                          ["Data da coleta","Nome da ETA / UTA", "Resultado", "week"]].reset_index(drop=True)

gee_vigi = pd.merge(vigi_selected, gee, on="week")
gee_vigi.to_csv("../../data/processed/ndvi_ndci_vigi.csv", index=False)
```

```python
df = pd.read_csv("../../data/processed/ndvi_ndci_vigi.csv")
```

```python
df_selected = df.loc[df["NDVI_median"] > -.9,
                     ["Data da coleta",
                      "date",
                      "NDVI_min",
                      "NDVI_max",
                      "NDVI_mean",
                      "NDVI_median",
                      "NDVI_stdDev",
                      "NDCI_min",
                      "NDCI_max",
                      "NDCI_mean",
                      "NDCI_median",
                      "NDCI_stdDev",
                      "Resultado"]]
```

```python
X = df_selected[["NDVI_min", "NDVI_max", "NDVI_mean", "NDVI_median", "NDVI_stdDev",
                "NDCI_min", "NDCI_max", "NDCI_mean", "NDCI_median", "NDCI_stdDev"]]
y = df_selected["Resultado"]
```

```python
import seaborn as sns

sns.pairplot(X)
```

```python
selected_vars = ["NDVI_median", "NDVI_max", "NDCI_median", "NDCI_max"]
X_selected = X[selected_vars]
```

## Regressão Linear

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import median_absolute_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import FunctionTransformer

RANDOM_STATE = 123

X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size = 0.2, random_state = RANDOM_STATE)
```

## 1. Regressao Linear

```python
lin_reg = LinearRegression()

transformer = FunctionTransformer(np.log1p, validate=True)
X_train_transf = transformer.transform(X_train)
lin_reg.fit(X_train_transf, y_train)
scores_lr = cross_val_score(lin_reg, X_train_transf, y_train, cv=5, scoring="neg_median_absolute_error")
```

```python
scores_lr
```

```python
median_absolute_error(lin_reg.predict(X_train_transf), y_train)
# Underfittando
```

```python
median_absolute_error(lin_reg.predict(transformer.transform(X_test)), y_test)
```

```python
import plotly.express as px

def plot_prediction(model, X_prep, y, cv_mae_scores, model_name):
    
    y_pred = model.predict(X_prep)

    fig = px.scatter(x=y_pred, y=y,
                    labels={
                         "x": "Valor previsto",
                         "y": "Valor real"
                     },
                    title=f"{model_name}: MAE médio CV5 = {round(-1 * cv_mae_scores.mean())}",
                    template="plotly_white")

    fig.update_layout(shapes = [{'type': 'line', 'yref': 'paper', 'xref': 'paper', 'y0': 0, 'y1': 1, 'x0': 0, 'x1': 1}])
    
    return fig
```

```python
plot_prediction(lin_reg, transformer.transform(X_selected), y, scores_lr, "Linear Regression")
```

## 2. XGBoost

```python
from xgboost import XGBRegressor

xgbr = XGBRegressor(max_depth=4, eta=.1, objective="reg:squarederror")
xgbr.fit(X_train, y_train, eval_metric="mae")
scores_xgb = cross_val_score(xgbr, X_train, y_train, cv=5, scoring="neg_median_absolute_error")
```

```python
from sklearn.model_selection import StratifiedKFold


xgbr = XGBRegressor(max_depth=4, eta=.1, objective="reg:squarederror")

params = {
        "eta": [0.1, 0.5, 0.7, 0.9],
        'min_child_weight': [1, 2, 3, 4, 5],
        "colsample_bytree": [0.6, 0.8, 1.0],
        "max_depth": [2, 3, 4, 5]
        }

from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(estimator=xgbr, param_grid=params,
                    scoring="neg_median_absolute_error",
                    cv=5, verbose=3)

grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
xgb_best = grid.best_estimator_
xgb_best.fit(X_train, y_train)

scores_xgbr = cross_val_score(xgb_best, X_train, y_train, cv=5, scoring="neg_median_absolute_error")

plot_prediction(xgb_best, X_selected, y, scores_xgbr, "XGBoost")
```

```python
median_absolute_error(xgb_best.predict(X_test), y_test)
```

```python
median_absolute_error(xgb_best.predict(X_train), y_train)
```

## 3. Comentários finais

Embora o modelo de XGBoost tenha obtido um menor MAE no CV e em teste quando em comparação com o modelo de regressão linear, embora tenha sido efetuado o train/test split e tunagem com cross-validation, verifica-se que o XGBoost apresenta um MAE de treino praticamente nulo, indicando overfitting. Entende-se que é um desafio inerente à regressão de valores extremos: na maior parte das vezes os valores são muito baixos, mas em pouca parte das instâncias haverão valores muito elevados que podem influenciar o modelo. A escolha da MAE como métrica tende a minimizar essa problemáticas, mas, como visto, não resolve o problema em sua totalidade.


Para superar esse limitação, seguindo as ideias de Geron (2019), poderíamos:
- Tentar um modelo mais simples
- Conseguir mais dados
- Reduzir o ruído nos dados
- Regularizar o modelo
